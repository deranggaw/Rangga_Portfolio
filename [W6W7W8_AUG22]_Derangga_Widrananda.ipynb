#### Saham

```python
dub_saham = df_dub1.copy()
list_columns = dub_saham.columns.to_list()
drop_columns = list_columns[5:]
dub_saham.drop(columns=drop_columns,axis = 1,inplace=True)
dub_saham
```

```python
# Hitung kuartil 1 dan 3
Q1 = df_dub1['Saham_AUM'].quantile(0.25)
Q3 = df_dub1['Saham_AUM'].quantile(0.75)

#Hitung interkuartil range
IQR = Q3 - Q1
boxplot_min = Q1 - 1.5 * IQR
boxplot_max = Q3 + 1.5 * IQR
print('Q1:\n',Q1)
print('\nQ3:\n',Q3)
print('\nIQR:\n',IQR)
print('\nMin:\n',boxplot_min)
print('\nMax:\n',boxplot_max)
```

```python
# Remove Outlier
filter_saham_min = dub_saham['Saham_AUM']<boxplot_min
filter_saham_max = dub_saham['Saham_AUM']>boxplot_max


df_saham_non_outlier = dub_saham[~
    (filter_saham_min | filter_saham_max)
    ]

df_saham_non_outlier
```
